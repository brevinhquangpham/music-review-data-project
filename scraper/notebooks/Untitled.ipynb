```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('../combined_data.csv')

# Replace every instance of $ with S in the 'Album_Name' column
df['Album_Name'] = df['Album_Name'].str.replace('$', 'S', regex=True)

# Extract the 'Average_Rating' column for the histogram
ratings = df['Average_Rating']

# Set the font and disable LaTeX
plt.rc('text', usetex=False)
plt.rcParams['font.family'] = 'Arial Unicode MS'

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(ratings, bins=20, alpha=0.7, edgecolor='black')

# Set the title and labels
plt.title('Distribution of Album Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
